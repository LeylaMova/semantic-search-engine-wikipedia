```python
class StemmerTokenizer(object):
    def __init__(self):
        self.ps = PorterStemmer()
        
    def __call__(self, doc):
        return [self.ps.stem(t) for t in word_tokenize(doc)]
    
cv = CountVectorizer(tokenizer=StemmerTokenizer(),stop_words='english',max_df=0.95, min_df=1)  
```

```python
sql = """
        select cate_vec.category_id, category.category_name, cate_vec.category_vec 
        from cate_vec 
        join category 
        on cate_vec.category_id = category.category_id
        order by cate_vec.category_vec desc;
        """
df_cat = pd.read_sql(sql, con=con, columns=['category_id','category_name','category_vec'])
```

```python
sql = """
        select page."page_id", page."page_vec", t1."category_name"
        from page_vec as page
        join (select page_cate."page_id", category."category_name"
              from page_cate
              join category
              on page_cate."category_id" = category."category_id") as t1         
        on t1."page_id" = page."page_id"
        order by page."page_vec" desc;
        """
df_page = pd.read_sql(sql, con=con)
```
